```python
from bandit import Bandit, EpsilonGreedy, ThompsonSampling, Visualization, comparison
```

## Initializing Constants

```python
Bandit_Reward = [1, 2, 3, 4]
num_trials = 20000
```

## Epsilon - Greedy Algorithm

```python
result = EpsilonGreedy(Bandit).experiment(Bandit_Reward, num_trials)
```

```python
EpsilonGreedy(Bandit).report(num_trials, result, algorithm = "EpsilonGreedy")
```

## Visualizing Learning Curve of Epsilon Greedy Algorithm

```python
Visualization().plot1(num_trials, result, 'Epsilon Greedy')
```

## Thomspon Sampling

```python
res1 = ThompsonSampling(Bandit).experiment(Bandit_Reward, num_trials)
```

```python
ThompsonSampling(Bandit).report(num_trials, res1, 'ThompsonSampling')
```

## Visualizing Learning Curve of Thompson Sampling

```python
Visualization().plot1( num_trials,res1, algorithm ='Thompson Sampling')
```

## Comparison of Epsilon Greedy and Thompson Sampling

### 1. Cumulative Regret and Reward

```python
Visualization().plot2(result, res1)
```

### 2. Cumulative Average Reward

```python
comparison(num_trials, result, res1)
```
